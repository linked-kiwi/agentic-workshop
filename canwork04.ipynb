We can now import the search function and set up a handle to run it.

```python
from duckduckgo_search import DDGS
quack = DDGS()
```

Let's run a query to check the current price of bitcoin.  We'll ask for the first three results.

```python
results = quack.text("Current price of bitcoin",max_results=3)
```

We'll get back a list of results from DuckDuckGo which include a title, URL, and a body containing the content.  Let's iterate through the results and display them.

```python
for item in results:
  print('\033[1m'+item['title']+"\033[0m] "+item['href'])
  print(item['body']+"\n")
```

We can also use the built in models to enable a simple chatbot directly via DucvkDuckGo. We can use a number of available models: claude-3-haiku, llama-3,3-70b, gpt-4o-mini, mistral-small-3; we'll use clause.

```python
results = quack.chat("Who played the first Dr Who?", model='claude-3-haiku')
print(results)
```

There's also a built in newsfeed in DuckDuckGo, which we can access. We'll select the Australian region and the top 10 news items for the last week in the Financial Times.

```python
results = quack.news(keywords="FT", region="au-en", safesearch="off", timelimit="w", max_results=20)
print(results)
for item in results:
  print('\033[1m'+item['date'][0:10]+' '+item['title']+'\033[0m ')
  print(item['body']+'\n')
```
